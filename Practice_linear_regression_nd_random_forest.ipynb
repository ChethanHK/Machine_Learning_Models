```python
import pandas as pd
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/master/delaney_solubility_with_descriptors.csv')
```

```python
df.head()
```

```python
###differentiat x nd y variable
```

```python
y=df[['logS']]
```

```python
x=df.drop('logS', axis=1)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=.80,random_state=42)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
##model building
```

```python
##linear regression
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
## applying the model to make_predictions
```

```python
y_train_pred=lr.predict(x_train)
```

```python
y_test_pred=lr.predict(x_test)
```

```python
print(y_train_pred)
```

```python
y_test_pred
```

```python
## model performance
```

```python
# module evalotion
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
train_mse=mean_squared_error(y_train,y_train_pred)
train_r2=r2_score(y_train,y_train_pred)
test_mse=mean_squared_error(y_test,y_test_pred)
test_r2=r2_score(y_test,y_test_pred)
```

```python
print('lr_mse (train)',train_mse)
print('lr_r2_score (train)',train_r2)
print('lr_mse (test)',test_mse)
print('lrr2_score (test)',test_r2)
```

```python
lr_predict=pd.DataFrame(['Linear_rgression',train_mse,train_r2,test_mse,test_r2]).transpose()
```

```python
lr_predict.columns=['method','training_mse','training_r2_score','test_mse','test_r2_score']
lr_predict
```

```python
##random forest
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf=RandomForestRegressor(max_depth=2,random_state=100)
```

```python
rf.fit(x_train,y_train)
```

```python
y_rf_train_pred=rf.predict(x_train)
y_rf_test_pred=rf.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
train_mse_rf=mean_squared_error(y_train,y_rf_train_pred)
train_r2_rf=r2_score(y_train,y_rf_train_pred)
test_mse_rf=mean_squared_error(y_test,y_rf_test_pred)
test_r2_rf=r2_score(y_test,y_rf_test_pred)
```

```python
rf_predict=pd.DataFrame(['Random_forest',train_mse_rf,train_r2_rf,test_mse_rf,test_r2_rf]).transpose()
```

```python
rf_predict.columns=['method','training_mse','training_r2_score','test_mse','test_r2_score']
rf_predict
```

```python
# model comparision
```

```python
dd=pd.concat([lr_predict,rf_predict],axis=0)
```

```python
dd=dd.reset_index(drop=True)
```
